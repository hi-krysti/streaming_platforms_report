```python
all_data = pd.concat([disney_data, hulu_data, netflix_data])
```

```python
filtered_data = all_data[all_data['release_year'] > 2000]
```

```python
count_after_2000 = filtered_data.shape[0]
```

```python
total_titles = all_data.shape[0]
```

```python
over_half_after_2000 = count_after_2000 > (total_titles / 2)
```

```python
## When was most content on the platforms released? ##
## Assumption was most content was released after the year 2000 ##

labels = ['Released After 2000', 'Released Before 2000']
sizes = [count_after_2000, total_titles - count_after_2000]
colors = ['lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Release Years')
plt.show()
```

```python
release_counts = filtered_data['release_year'].value_counts().sort_index()
```

```python
## Number of releases for years after 2000 ##

plt.figure(figsize=(12, 6))
sns.barplot(x=release_counts.index, y=release_counts.values, color='skyblue')
plt.title('Number of Releases for Each Year (After 2000)')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```

```python
release_counts = all_data['release_year'].value_counts().sort_index()
```

```python
## Total releases for each year for all platforms ##

plt.figure(figsize=(20, 6))
sns.barplot(x=release_counts.index, y=release_counts.values, color='lightgreen')  
plt.title('Number of Releases for Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=90)
plt.show()

## The approach we took was to look at the data in a few different ways to really understand when the bulk of it was released. This last chart expands on that and shows that from the data, 2019 had the most releases overall ##
```

```python
## Which streaming platform has the most 'shows', which has the longest-running shows (most seasons) ##
## Number of seasons for Disney shows, assumption was Disney had the most long-running shows ##

tv_shows_data = disney_data[disney_data['type'] == 'TV Show'].copy()  


tv_shows_data['num_seasons'] = tv_shows_data['duration'].str.extract('(\d+)').astype(float)


plt.figure(figsize=(10, 6))
sns.countplot(x='num_seasons', data=tv_shows_data, palette='viridis')  
plt.title('Number of Seasons for Disney+ TV Shows')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.show()
```

```python
## Number of seasons for all streaming platforms, which one has the most 'TV Shows' ##
## Which platform has TV Shows with the most seasons ##

hulu_data['platform'] = 'Hulu'
disney_data['platform'] = 'Disney+'
netflix_data['platform'] = 'Netflix'
amazon_data['platform'] = 'Amazon Prime'

disney_tv_shows_data = disney_data[disney_data['type'] == 'TV Show'].copy()
disney_tv_shows_data['num_seasons'] = disney_tv_shows_data['duration'].str.extract('(\d+)').astype(float)

netflix_tv_shows_data = netflix_data[netflix_data['type'] == 'TV Show'].copy()
netflix_tv_shows_data['num_seasons'] = netflix_tv_shows_data['duration'].str.extract('(\d+)').astype(float)

hulu_tv_shows_data = hulu_data[hulu_data['type'] == 'TV Show'].copy()
hulu_tv_shows_data['num_seasons'] = hulu_tv_shows_data['duration'].str.extract('(\d+)').astype(float)

amazon_tv_shows_data = amazon_data[amazon_data['type'] == 'TV Show'].copy()
amazon_tv_shows_data['num_seasons'] = amazon_tv_shows_data['duration'].str.extract('(\d+)').astype(float)

all_tv_shows_data = pd.concat([disney_tv_shows_data, netflix_tv_shows_data, hulu_tv_shows_data, amazon_tv_shows_data],
                              ignore_index=True)

plt.figure(figsize=(12, 8))
sns.countplot(x='num_seasons', hue='platform', data=all_tv_shows_data, palette='viridis')
plt.title('Number of Seasons for TV Shows on Streaming Platforms')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='platform', data=all_tv_shows_data, palette='viridis')
plt.title('Number of TV Shows for Each Streaming Platform')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.show()

## We again took the approach of looking at the data and expanding on it in a few ways. Our initial assumption was wrong, Netflix has the most shows and the longest running shows ##
```

```python
## Which month is most common for content to be released? ## 
## Assumption is Sep and Oct are the most popular release months ##

hulu_data['platform'] = 'Hulu'
disney_data['platform'] = 'Disney+'
netflix_data['platform'] = 'Netflix'
amazon_data['platform'] = 'Amazon Prime'

all_data = pd.concat([hulu_data, disney_data, netflix_data, amazon_data], ignore_index=True)

all_data['date_added'] = pd.to_datetime(all_data['date_added'], errors='coerce')

all_data['release_month'] = all_data['date_added'].dt.month_name()

plt.figure(figsize=(12, 8))
sns.countplot(x='release_month', hue='platform', data=all_data, palette='viridis', dodge=True)
plt.title('Number of Releases by Month for Streaming Platforms')
plt.xlabel('Release Month')
plt.ylabel('Number of Releases')
plt.show()
```

```python
## Looking at the trends of release months over the years to expand on the data ##
all_data['release_year'] = all_data['date_added'].dt.year

plt.figure(figsize=(15, 8))
sns.countplot(x='release_month', hue='release_year', data=all_data, palette='deep', dodge=True)
plt.title('Number of Releases by Month and Year for Streaming Platforms')
plt.xlabel('Release Month')
plt.ylabel('Number of Releases')
plt.legend(title='Release Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
## We wanted a clearer visualization to show the month with most releases ##
counts = all_data['release_month'].value_counts()

fig, ax = plt.subplots(figsize=(15, 8))

ax.bar(counts.index, counts, color='skyblue', edgecolor='black')

ax.set_xlabel('Release Month')

ax.set_ylabel('Number of Releases')

ax.set_title('Histogram of Releases by Month for Streaming Platforms')

plt.show()

## Again our assumption was wrong, November is by far the most popular month to release content across all platforms ##
```
